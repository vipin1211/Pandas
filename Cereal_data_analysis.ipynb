```python
cut = pd.cut(cereals_data3.Cal, bins=3, labels=["low", "moderate", "high"])
cut.value_counts()
```

```python
qcut = pd.qcut(cereals_data3.Cal, q=3, labels=["low", "moderate", "high"])
qcut.value_counts()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 2, figsize = (9, 5))
ax[0].bar(["low", "moderate", "high"],cut.value_counts())
ax[0].set_title('Cut')
ax[0].set_xlabel('bins')
ax[0].set_ylabel('Count')
ax[1].bar(["low", "moderate", "high"],qcut.value_counts())
ax[1].set_title('Qcut')
ax[1].set_xlabel('bins')
ax[1].set_ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
'''
The major distinction is that qcut calculates the size of each bin in order to make sure the distribution of data in 
the bins is equal. In other words, all bins will have (roughly) the same number of observations but the bin range will 
vary.On the other hand, cut is used to specifically define the bin edges. There is no guarantee about the 
distribution of items in each bin.
'''
```

```python
import pandas as pd
def standardize(df,args):
    my_df  = pd.DataFrame(columns = args)
    for i in args:
        my_df[i]=(df[i]-df[i].mean())/df[i].std()
    return my_df
```

```python
cols=['Carbo', 'Sugars', 'Potass', 'Vit']
my_df=standardize(cereals_data3,cols)
```

```python
import random
def find_max(x, y, n_iter):
    x_current=random.choice(x)
    for i in range(n_iter):
        cur_ind=x.index(x_current)
        x_previous=x[cur_ind-1]
        x_next=x[cur_ind+1]
        f_current=y[cur_ind]
        f_previous=y[cur_ind-1]
        f_next=y[cur_ind+1]
        if f_next>f_previous:
            x_current= x_next
        else:
            x_current= x_previous
    print (x_current,f_current)
```

```python
find_max(list(x), y, 100)
```
